```python
from sklearn.cluster import KMeans
from collections import Counter
from matplotlib import pyplot as plt
import cv2
from sklearn.cluster import KMeans
from collections import Counter
from matplotlib import pyplot as plt
import cv2
%matplotlib inline
import numpy as np 
from skimage import io
from skimage import feature
from skimage.color import rgb2gray
import os
from glob import glob
import numpy as np
from skimage import morphology as mp
from skimage import filters as fl
```

```python
#finding last or first pixeleted value
def gety(img,t,sha):
    
    x=sha[0]
    y=sha[1]
    if(t=="f"):
        for i in range(0,x):
            for j in range(0,y):
                if(img[i][j]==1):
                    return i
    if(t=="l"):
        for i in range(x-1,0,-1):
            for j in range(0,y):
                if(img[i][j]==1):
                    return i   
                
def getx(img,t,sha):
    
    x=sha[0]
    y=sha[1]
    if(t=="f"):
        for i in range(0,y):
            for j in range(0,x):
                if(img[j][i]==1):
                    return i
    if(t=="l"):
        for i in range(y-1,0,-1):
            for j in range(0,x):
                if(img[j][i]==1):
                    return i                 
```

```python
def printimage(st,arr):
    x,y,w,h = [arr[0],arr[2],arr[1]-arr[0],arr[3]-arr[2]]
    img = cv2.imread(st)
    a=cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),5)
    cv2.imwrite( "./output/"+st, a );
```

```python
def getcord(st):
    x1=io.imread(st)
    #case 4 use of gaussian before rgb case 5 use of gaussian after canny
    x1=fl.gaussian(x1)
    x2=rgb2gray(x1)
    
    #case1
    #x3 = feature.canny(x2, .5, 0.5, 0.7)
    
    #case2
    """
    x3 = feature.canny(x2,1,.9,.8)
    x3=fl.gaussian(x3)
    x3=mp.binary_erosion(x3)
    """
    
    #case3
    x3 = feature.canny(x2,0.5,.8,0)
    #.9,.7,.8
    
    x3=fl.gaussian(x3)
    x3=mp.binary_erosion(x3)
    
    sha=x3.shape
    arr=[getx(x3,"f",sha),getx(x3,"l",sha),gety(x3,"f",sha),gety(x3,"l",sha)]
    printimage(st,arr)
    #return(arr)
    
    
```

```python
def load_images(base_path):
    images = []
    path = os.path.join(base_path, '*.png')
    #print(path)
    for i in glob(path):
        print(i)
        getcord(i)
```
