```python
#finding last pixeleted value
def gety(img,t,sha):
    
    x=sha[0]
    y=sha[1]
    if(t=="f"):
        for i in range(0,x):
            for j in range(0,y):
                if(img[i][j]==1):
                    return i
    if(t=="l"):
        for i in range(x-1,0,-1):
            for j in range(0,y):
                if(img[i][j]==1):
                    return i            
```

```python
#finding last pixeleted value
def getx(img,t,sha):
    
    x=sha[0]
    y=sha[1]
    if(t=="f"):
        for i in range(0,y):
            for j in range(0,x):
                if(img[j][i]==1):
                    return i
    if(t=="l"):
        for i in range(y-1,0,-1):
            for j in range(0,x):
                if(img[j][i]==1):
                    return i 
```
